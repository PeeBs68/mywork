# Probability Density Function - PDF

**Normal PDF**

$$ f(x) = \frac{1}{\sqrt{2 \pi}} e^{-\frac{1}{2} \big( \frac{x - \mu}{\sigma} \big)^2}$$


**Standard Normal PDF**

$$ f(x) = \frac{1}{\sqrt{2 \pi}} e^{-\frac{x^2}{2}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    a = 1.0 / np.sqrt(2.0 * np.pi)
    b = np.exp(-(x**2) / 2.0)
    return a*b
    
```

```python
# Test a value in the function

f(0.0)
```

```python
# Plot of x values for plotting

x = np.linspace(-5, 5.0, 1001)

# create empty plot
fig, ax = plt.subplots()

# Plot f(x)
ax.plot(x, f(x))
```

```python
# Plot of x values for plotting

x = np.linspace(-5, 5.0, 1001)

# create empty plot
fig, ax = plt.subplots()

#Plot histogram of randomly generated normal values from plotting_normal_dist.ipynb
rng = np.random.default_rng()
rand_normal = rng.standard_normal(10000)
ax.hist(rand_normal, bins=100, density=True);

# Plot f(x)
ax.plot(x, f(x))

```
