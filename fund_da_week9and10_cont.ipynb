## plots of different functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Range of values

x = np.linspace(-4.0, 4.0, 801)
```

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label = 'x**2')
ax.plot(x, x**2 + 1, label = 'x**2 + 1')
ax.plot(x, x**2 - 1, label = 'x**2 - 1')
ax.plot(x, x**2 - 3, label = 'x**2 - 3')
ax.legend()
```

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label = 'x**2')
ax.plot(x, x**2 + x, label = 'x**2 + x')
ax.plot(x, x**2 - x, label = 'x**2 - x')
ax.plot(x, x**2 - (4 * x), label = 'x**2 - 4x')
ax.legend()
```

```python
x = np.linspace(-12.0, 12.0, 801)

fig, ax = plt.subplots()

ax.plot(x, x**2, label = 'x**2')
ax.plot(x, x**2 + x, label = 'x**2 + x')
ax.plot(x, x**2 - x, label = 'x**2 - x')
ax.plot(x, x**2 - (4 * x), label = 'x**2 - 4x')
ax.legend()
```

```python
# PLotting a log value

x = np.linspace(0.0000001, 10, 1001)

fig, ax = plt.subplots()

ax.plot(x, np.log(x), label = 'log x')

ax.legend()
```

```python
# PLotting a log value

x = np.linspace(1, 10, 1001)

fig, ax = plt.subplots()

ax.plot(x, np.log(x), label = 'log x')
ax.plot(x, np.log2(x), label = 'log2 x')
ax.plot(x, np.log10(x), label = 'log10 x')

ax.legend()
```

```python
# PLotting a log value

x = np.linspace(-3.0, 3.0, 1001)

fig, ax = plt.subplots()

ax.plot(x, x, label = ' x')
ax.plot(x, x**2, label = 'x**2')
ax.plot(x, x**3, label = 'x**3')
ax.plot(x, x**4, label = 'x**4')

ax.legend()
```

```python
# PLotting more values

x = np.linspace(0.0, 50.0, 1001)

fig, ax = plt.subplots()

ax.plot(x, x, label = ' x')
ax.plot(x, x**2, label = 'x**2')
ax.plot(x, 2**x, label = '2**x')

ax.legend()
```
