### Using loc and iloc
loc uses labels (i.e. rows/columns names/numbers)   
iloc uses position

```python
df.loc[2:6]
```

```python
df.loc[:,'species']
```

```python
df.loc[:,['sepal_length', 'species']]
```

```python
# Creates a new df
# Note how the loc keeps the original position rather than starting at 0
newdf = df.loc[2:6, :]

newdf
```

```python
df.iloc[3]
```

```python
df.iloc[2:9, 3]
```

```python
# To get every second row

df.iloc[1:12:2, 3]
```

```python
df.at[2,'species']
```

### Using boolean selection

```python
df.loc[:,'species'] == 'setosa'
```

```python
x = df.loc[df.loc[:,'species'] == 'versicolor']

x
```

```python
x.loc[50,:]
```

Note how loc and iloc work diffeently in terms of the label/index

```python
x.iloc[0,:]
```

```python
df.loc[df.loc[:,'petal_width'] >= 1.6]
```
