# Plotting Histograms




```python
# generate a list of random numbers
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng()

#rng.random(10)
```

```python
# Random Floating Point Numbers

size = 1000
rand_floats = rng.random(size=size)

#rand_floats
```

```python
# Create an empty plot

fig, ax = plt.subplots()

# Plot a histogram

ax.hist(rand_floats);
```
