# Normal Looking Data

```python
# Generate Numbers on a Standard Normal Distribution

import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng()

#rng.random()

# 'standard' indicates the vast majority of values will be centered around 0. 
# sigma/deviations ect
rand_normal = rng.standard_normal(10000)



```

```python
fig, ax = plt.subplots()

# Plot a histogram

ax.hist(rand_normal, bins=100, density=True);
```
