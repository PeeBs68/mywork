## Plots

```python
# Plots
import matplotlib.pyplot as plt
```

```python
# Numerical Arrays
import numpy as np
```

Define a function for

$f(x) = x^2 - (4 * x) + 4$

```python
def f(x):
    return x**2 - (4 * x) - 4
```

```python
# Generating values

# Basic values manually entered into array
# x = np.array([-1, 0, 1, 2, 3, 4, 5, 6])

# Using arange to get list of values
x = np.arange(-1.0, 5.0, 0.01)

# Using linspace to get list of values
x = np.linspace(-1.0, 5.0, 601)
```

```python
# Calculate y values
y = f(x)
y
```

```python
# Create a plot
fig, ax = plt.subplots()

# Plot x v y

ax.plot(x, y)
```

```python
# Create a plot
fig, ax = plt.subplots()

# Plot x v y

#ax.plot(x, y, '.')
ax.plot(x, y)
```
